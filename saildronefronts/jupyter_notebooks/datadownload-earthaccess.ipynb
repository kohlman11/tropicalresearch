# MURSST

```python
import earthaccess

MUR_SST = earthaccess.search_data(
    short_name="MUR25-JPL-L4-GLOB-v04.2",
    temporal=("2017-12-06", "2017-12-08"),
)

earthaccess.download(MUR_SST, "./local_folder")


results = earthaccess.search_data(
    short_name="M2T1NXSLV",
    bounding_box=(-120, 30, -110, 40),  # Example bounding box for Seattle area
    temporal=("2024-06-01", "2024-07-01"),
    count=5
)

files = earthaccess.download(results, "./merra2_data")
```

# ERA5

```python
import cdsapi
import xarray as xr

# Assuming you've already obtained and configured your CDS API key

c = cdsapi.Client()

# This is a placeholder; replace with the actual request copied from the CDS website
c.retrieve(
    'reanalysis-era5-single-levels',
    {
        'product_type': 'reanalysis',
        'format': 'netcdf',
        'variable': [
            '10m_u_component_of_wind', '10m_v_component_of_wind',
        ],
        'year': '2020',
        'month': '01',
        'day': '01',
        'time': '12:00',
        'area': [
            50, -10, 40, 2,  # Example: North, West, South, East
        ],
    },
    'era5_wind_data.nc')

# After the download, open and work with the NetCDF file
ds = xr.open_dataset('era5_wind_data.nc')
```
